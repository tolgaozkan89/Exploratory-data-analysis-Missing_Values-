## aşağıda ise NA olan veriler için içinde bulunduğu sütunların yıllara göre ortalaması verilmiştir. 

```python
education2_df = education_df.copy()
years = education2_df["YEAR"].unique()

for col in fill_list:
    for year in years:
        YIL_SART = education2_df["YEAR"] == year
        education2_df.loc[YIL_SART, col] = education2_df.loc[YIL_SART, col].fillna(education2_df[YIL_SART][col].mean())

education2_df
```
